```python
#displaying Historgrams for big data.
#If you set bins to an integer, say bins=10, Matplotlib will divide your data range into 10 equally spaced bins.
import numpy
import matplotlib.pyplot as plt
import random
x=numpy.random.uniform(1,100,10)
y=[random.randint(1,100) for _ in range(10)]
print(y)
plt.hist(x,bins=5,color='red')
plt.title("Histograms shows details about numbers persons for each sectors :")
plt.xlabel("values")
plt.ylabel("Frequency")
plt.show()
```

```python
#displaying Historgrams for big data.
#If you set bins to an integer, say bins=10, Matplotlib will divide your data range into 10 equally spaced bins.
import numpy
import matplotlib.pyplot as plt
import random
x=numpy.random.uniform(1,100,10)
y=[random.randint(1,100) for _ in range(10)]
print(y)
plt.hist(x,bins=5,color='red',edgecolor="black")
plt.title("Histograms shows details about numbers persons for each sectors :")
plt.xlabel("values")
plt.ylabel("Frequency")
plt.show()
```

```python
#displaying Historgrams for big data.
#If you set bins to an integer, say bins=10, Matplotlib will divide your data range into 10 equally spaced bins.
import numpy
import matplotlib.pyplot as plt
import random
x=numpy.random.uniform(1,100,10)
y=[random.randint(1,100) for _ in range(10)]
print(y)
plt.hist(x,bins=5,color='red',edgecolor="black")
plt.title("Histograms shows details about numbers persons for each sectors :")
plt.xlabel("values")
plt.ylabel("Frequency")
plt.grid(True, linestyle='-', alpha=0.5)
plt.legend(['Ages'])
plt.show()
```
