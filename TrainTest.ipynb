```python
import numpy
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
import random
numpy.random.seed(2)
x = numpy.random.normal(3, 1, 100)
y = numpy.random.normal(150, 40, 100) / x
plt.scatter(x, y)
plt.legend(['Total Data'])
plt.show()


#Trained data


train_x=x[:80]
train_y=y[:80]
plt.scatter(train_x,train_y)
plt.legend(['Trained Data'])
plt.xlabel("Time")
plt.ylabel("Money in rupees")
plt.show()

#Testing data


test_x=x[80:]
test_y=y[80:]
plt.scatter(test_x,test_y)
plt.legend(['Testing Data'])
plt.xlabel("Time")
plt.ylabel("Money in rupees")
plt.show()

# Building polynomial model 
# Why polynomial regression, because polynomial regreession covers the all the points


model=numpy.poly1d(numpy.polyfit(train_x,train_y,4))
#0 and 6 indiactes starting and ending values on X axis.That's the range of X-axis
#25 indicates the how many points should be used to build the line
line=numpy.linspace(0,6,25)
plt.scatter(train_x,train_y)
plt.plot(line,model(line))
plt.xlabel("Time")
plt.ylabel("Money in rupees")
plt.scatter(line,model(line))
plt.legend(['Trained data','Predicted line','Predicted values'])
plt.show()

#Now calculating r2 value. Let's see how good data set is fitting into model
#It measures the relationship between the x axis and the y axis,
#and the value ranges from 0 to 1, where 0 means no relationship, and 1 means totally related.


trained_r2=r2_score(train_y,model(train_x)) # 0.7343432


#since r2 value is nearly good that means given dataset is fitting good into model
print("R-squared value :",trained_r2)


#Now we want to test the model with the testing data as well, to see if gives us the same result.
#we just gonna check the R squared value of model by giving train sets to model


test_r2=r2_score(test_y,model(test_x))
print("R-squared Value : ",test_r2)

#R-squared value : 0.7988645544629793
#R-squared Value :  0.8086921460343643
# Since the both R-squared values are somewhat realted to model.
#Note: The result 0.809 shows that the model fits the testing set as well, 
#and we are confident that we can use the model to predict future values.

# Now we're gonna Predict the values using model

time_taken=eval(input("Taken time :"))
money_spent=model(time_taken)
print("{} rupees money spent on {} minutes of  time.".format(money_spent,time_taken))
plt.scatter(time_taken,money_spent,c='red',s=250)


#plt.annotate(f"({time_taken}, {money_spent})", (time_taken,money_spent), textcoords="offset points", xytext=(0,10), ha='left')


plt.annotate(f"({time_taken}, {money_spent})", (time_taken,money_spent), textcoords="offset points", xytext=(0,10), ha='left')
plt.xlabel("Time")
plt.ylabel("Money in rupees")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Plot the points using scatter
plt.scatter(x, y, color='blue')

# Annotate each point with its value
for i in range(len(x)):
    plt.annotate(f"({x[i]}, {y[i]})", (x[i], y[i]))  # Displaying (x, y) values as text

# Add labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot with Annotated Points')

# Display the plot
plt.grid(True)
plt.show()
```
