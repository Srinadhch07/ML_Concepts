```python
#Working with linear Regression
from scipy import stats
import matplotlib.pyplot as plt
#Let x and y be the inputs datasets to linear regression
x=[1,3]
y=[3,2]
regression=stats.linregress(x,y)
print("Slope :",regression.slope)
print("Interceptor :",regression.intercept)
print("R- Value :",regression.rvalue)
print("P- Value :",regression.pvalue)
print("Standard Error :",regression.stderr)
plt.scatter(x,y)
plt.show()
```

```python
#Working with linear Regression
from scipy import stats
import matplotlib.pyplot as plt
#Let x and y be the inputs datasets to linear regression
x=[3,4]
y=[2,3]
regression=stats.linregress(x,y)
print("Slope :",regression.slope)
print("Interceptor :",regression.intercept)
print("R- Value :",regression.rvalue)
print("P- Value :",regression.pvalue)
print("Standard Error :",regression.stderr)

# The below function is used to represent the linear regression in the graph.
def fun(x):
    return regression.slope*x+regression.intercept # Represents regression line.
linePoints=list(map(fun,x))
print("Line Regression Points :",linePoints)
plt.scatter(x,y)
plt.plot(x,linePoints)
plt.show()
```

```python
#Working with linear Regression 
#Trial 5

# Goal is to predict the person height based on persons weight
from scipy import stats
import matplotlib.pyplot as plt
import random
#Let x and y be the inputs datasets to linear regression
x=[random.randint(1,100) for _ in range(1,10)]
y=[random.randint(1,100) for _ in range(1,10)]
print("x =",x,"\ny =",y)
regression=stats.linregress(x,y)
print("Slope :",regression.slope)
print("Interceptor :",regression.intercept)
print("R- Value :",regression.rvalue)
print("P- Value :",regression.pvalue)
print("Standard Error :",regression.stderr)

# The below function is used to represent the linear regression in the graph.
def fun(x):
    return regression.slope*x+regression.intercept # Represents regression line.
linePoints=list(map(fun,x))
print("Line Regression Points :",linePoints)
plt.xlabel("Person Weight")
plt.ylabel("Person Height")
plt.scatter(x,y)
plt.plot(x,linePoints)
plt.show()
```
