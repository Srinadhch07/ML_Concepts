```python
#Scatter plot is one the method provided by matplotlib library in python
# scatter plot is represents the data with dots in the graphs
#scatter plot graph is needed two input data sets. i.e ., X and Y
import matplotlib.pyplot as plt 
import random
height=[random.randint(1,100) for _ in range(10)]
weight=[random.randint(1,100) for _ in range(10)]
plt.scatter(height,weight)
plt.xlabel('Weights of persons ')
plt.ylabel('Heights of persons ')
plt.show()
```

```python
#Scatter plot is one the method provided by matplotlib library in python
# scatter plot is represents the data with dots in the graphs
#scatter plot graph is needed two input data sets. i.e ., X and Y
import matplotlib.pyplot as plt 
import random
height=[random.randint(1,100) for _ in range(5)]
weight=[random.randint(1,100) for _ in range(5)]
sizes=[250,100,400,250,320]

plt.scatter(height,weight,s=sizes) #s defines the size of the of each plotting point data in datesets.
# Note : array of sizes nust be equal to sizes of X and Y
plt.xlabel('Weights of persons ')
plt.ylabel('Heights of persons ')
plt.show()
```

```python
# color attribute..............
#Scatter plot is one the method provided by matplotlib library in python
# scatter plot is represents the data with dots in the graphs
#scatter plot graph is needed two input data sets. i.e ., X and Y
import matplotlib.pyplot as plt 
import random
height=[random.randint(1,100) for _ in range(5)]
weight=[random.randint(1,100) for _ in range(5)]
sizes=[250,100,400,250,320]

plt.scatter(height,weight,s=sizes, c='red') #s defines the size of the of each plotting point data in datesets.
# Note : array of sizes nust be equal to sizes of X and Y
plt.xlabel('Weights of persons ')
plt.ylabel('Heights of persons ')
plt.show()
```

```python
# Alpha attribute in scatter method
#Scatter plot is one the method provided by matplotlib library in python
# scatter plot is represents the data with dots in the graphs
#scatter plot graph is needed two input data sets. i.e ., X and Y
import matplotlib.pyplot as plt 
import random
height=[random.randint(1,100) for _ in range(5)]
weight=[random.randint(1,100) for _ in range(5)]
sizes=[250,100,400,250,320]

#alpha attribute used to for transperency.And ranged from 0.0 to 1.0 only
plt.scatter(height,weight,s=sizes, c='red',alpha=0.1) 

plt.xlabel('Weights of persons ')
plt.ylabel('Heights of persons ')
plt.show()
```
