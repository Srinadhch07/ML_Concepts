```python
import pandas
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier

#Reading CSV file using pandas
data=pandas.read_csv("comedy.csv")
print(data)

#Since we need to conevrts all non-intergers values into integer values.Hence we assign values to the non-integer attribute values
mapping={'UK':0,'USA':1,'N':2}
#Now, we Gonna map the values to contraints
data['Nationality']=data['Nationality'].map(mapping)
mapping={'YES':1,'NO':2}
data['Go']=data['Go'].map(mapping)
constraints=['Age','Experience','Rank','Nationality']

#Listing all contraints attributes,we doing this to fit all the contraints attributes into Tree model
# This contraints attributes value  acts as X values in Tree model
contraint_data=data[constraints]
#This result_data acts as Y value in TreeModel
Result_data=data['Go']

#Building DecisionTreeClassifer model
TreeModel=DecisionTreeClassifier()

#Fitting the constraints and result attributes into the model
#These Constaints attributes and Result attributes are neccessary to train the decsion tree

TreeModel=TreeModel.fit(contraint_data,Result_data)

#PLotting the Trianed Decision Tree
tree.plot_tree(TreeModel,feature_names=constraints)

#Since, Till now we have trianed the decision tree.Now we Gonna use the Decision tree to predict the values.
# To predict the Values, we need to provide available data i.e., we  provide available contraints to model
print("Predicted value : ",TreeModel.predict([[40, 10, 7, 1]]))


'''
#Explaination of DesionTree Output

* Rank :
    if the rank is less than or equal 6.5 then True arrow (Left arrow) otherwise False arrow (Right Arrow)
* Gini :
    Gives the quality of split
    it retuns the nomber between the 0.0 and 0.5
    if gini=0.0, that means all the sample got same result
    if gini=0.5, that means split done at exactly middle

    #### There are many ways to split the datasets, We use gini method to split.
    #####

        Formula : 
        
        Gini = 1 - (x/n)2 - (y/n)2
        
    #####
* Samples:
    samples returns the numbers of sample remained at present desicion
    # That means at particular decision, it returns numbers of samples remained to check.
* Values:
    syntax : values[NO,YES]
        values represents the how many samples goes to NO and how many samples goes to YES.
'''
```

```python
import pandas
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt

df = pandas.read_csv("comedy.csv")

d = {'UK': 0, 'USA': 1, 'N': 2}
df['Nationality'] = df['Nationality'].map(d)
d = {'YES': 1, 'NO': 0}
df['Go'] = df['Go'].map(d)

features = ['Age', 'Experience', 'Rank', 'Nationality']

X = df[features]
y = df['Go']

dtree = DecisionTreeClassifier()
dtree = dtree.fit(X, y)

tree.plot_tree(dtree, feature_names=features)
```
