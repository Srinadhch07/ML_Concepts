```python
#Working with Multiple regression
# Graph Representaions
import pandas # inmporting pandas to read the csv file
from sklearn import linear_model # we are importing linear regression directly from sklearn 
import matplotlib.pyplot as plt

df=pandas.read_csv("data.csv") # reading CSV file for datasets

X = df[['Weight', 'Volume']] # reading the weight and Volume from csv file.
'''
In the above line we setting up the datasets which is used to train the model
'''
y=df['CO2'] # reading the previous data which we have to predict
regr = linear_model.LinearRegression() # Building linear regression model to predict the values
'''
In the above linear model even it is linear model , we trainninf it with multiple data sets.
i.e., for X we have given weight and volume
and for Y have given CO2
'''
regr.fit(X, y) # Ftting the Data into model
plt.scatter(df['Weight'],df['Volume'])
plt.xlabel('Weight')
plt.ylabel('Volume')
plt.legend(['Trained Data(Previous Data)'])
plt.show()
# Finding the coefficients of Model
print(regr.coef_)


weight=int(input("Enter Weight"))
volume=int(input("Enter Volume"))
predictedCO2 = regr.predict([[weight, volume]]) # predcting the CO2 values based on given Wieght and Volume
print("Predicted CO2 Emisssion :",predictedCO2)
'''
To represent the Predicted values on Graph in multiple regression
we can't provide weight and height at X axis simultaneously Soo
for convenience , here in this graph using weight as x axis.
The X axis can also be Volume
'''
plt.scatter(weight,predictedCO2,color='red')
plt.xlabel('Weight')
plt.ylabel('CO2')
plt.legend(['Predicted CO2 emission.'])
plt.show()


"""
[0.00755095 0.00780526]

What is use of Coeffients ?
The answer is simple
These values tell us that if the weight increase by 1kg, the CO2 emission increases by 0.00755095g.

And if the engine size (Volume) increases by 1 cm3, the CO2 emission increases by 0.00780526 g.

whenever if input values varries.It describes  how much change per 1 unit.

if weight incresed by 1 , co2 emmission increased by 0.00755095

0.00755095 x**2 + 0.00780526 x = CO2

"""
```
